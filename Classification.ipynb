## Load data

```python
#############################################################
#         TODO: read dataset as pandas dataframe            #
#############################################################

df = pd.read_excel('titanic.xls')
```

```python
#############################################################
#            TODO: Convert the dataset to X,y               #
#############################################################

X = df.drop('survived', axis=1)
y = df['survived']
```

```python
X.isna().sum()
```

## Preprocessing

```python
########################################################################
#     TODO: Delete columns that are not used or have too many nulls    #
########################################################################

X = X.drop(['name', 'ticket', 'cabin', 'boat', 'body', 'home.dest'], axis=1)
```

```python
X.head()
```

```python
X.nunique()
```

```python
X.isna().sum()
```

```python
############################################################################################
#     TODO: Fill in the nulls with the appropriate value, such as median, mean or mode     #
############################################################################################

median_age = X['age'].median()
median_fare = X['fare'].median()

X.fillna({'age': median_age, 'fare': median_fare}, inplace=True)

mode_embarked = X['embarked'].mode()[0]
X['embarked'].fillna(mode_embarked, inplace=True)
```

```python
# All columns must not contain null

X.isna().sum()
```

```python
X.nunique()
```

```python
#############################################################
#    TODO: create Dummy Variable for categorical columns    #
#############################################################

X = pd.get_dummies(X, columns=['sex', 'embarked'], prefix=['is', 'embarked'], drop_first=True)
```

```python
df.info()
```

```python
#############################################################
#              TODO: Normalize numeric columns              #
#############################################################
from sklearn.preprocessing import StandardScaler

cols_to_normalize = ['age', 'sibsp', 'parch', 'fare']
scaler = StandardScaler().fit_transform(X[cols_to_normalize])
X[cols_to_normalize] = scaler
```

```python
from sklearn.decomposition import PCA

##############################################################
#  TODO: Reduce input dimensions to 5 dimensions using PCA   #
##############################################################

n_components = 5
X_pca = PCA(n_components=n_components).fit_transform(X)
X_pca_df = pd.DataFrame(X_pca, columns=[f"PC{i}" for i in range(1, n_components+1)])
```

```python
#############################################################
#              TODO: split to train and test                #
#############################################################

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca_df, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Desisiom Tree

```python
from sklearn import tree
###################################################################
#  TODO: Train the decision tree classifier on the training data  #
###################################################################
clftree = tree.DecisionTreeClassifier(random_state=0)
clftree.fit(X_train, y_train)
```

### Predict values using trained model

```python
#####################################################################
# TODO: Predict the output class for the training and test datasets #
#####################################################################

y_train_pred =  clftree.predict(X_train)
y_test_pred =  clftree.predict(X_test)
```

### evaluation metrics

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score
#####################################################################
# TODO: Calculate accuracy for training and all metrics for testing #
#####################################################################
train_acc, test_acc, test_precision, test_recall, test_f1_score, test_conf_matrix = [accuracy_score(y_train, y_train_pred),
                                                                                   accuracy_score(y_test, y_test_pred),
                                                                                   precision_score(y_test, y_test_pred),
                                                                                   recall_score(y_test, y_test_pred),
                                                                                   f1_score(y_test, y_test_pred),
                                                                                   confusion_matrix(y_test, y_test_pred)]

print("Training accuracy:", train_acc)
print("Testing accuracy:", test_acc)
print("Testing precision:", test_precision)
print("Testing recall:", test_recall)
print("Testing F1 score:", test_f1_score)
print("Testing confusion matrix:")
print(test_conf_matrix)
```
